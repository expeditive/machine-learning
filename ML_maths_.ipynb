```python
print(a_dot_b) #dot product is always a scalar
```

```python
print(c_cross_d)
```

projection of "a" vector on "v" vector

---
proj on a = ((a.v) / |v|^2 )*v



```python
a = np.asarray([2,5])

v = np.asarray([8,-6])

#magnitude of v vector
mag_v = np.sqrt(np.sum(v**2))


proj_of_a_on_v = (np.dot(a,v)/mag_v**2)*v

print('projection of vector a on v', proj_of_a_on_v)
```

matrix addition --> both matrix must have the same shape

```python
A = np.random.randint(90, size = (3,4))

print("A = \n",A)

B = np.random.randint(45,size = (3,4))

print("B = \n",B)

sum = A + B

print('A+B = \n',sum)
```

```python
A = np.random.randint(90, size = (3,4))

print("A = \n",A)

B = np.random.randint(45,size = (3,4))

print("B = \n",B)

diff = A - B

print('A-B = \n',diff)
```

```python
#can alos add by add function
sum2 = np.add(A,B)
print(sum2)
```

```python
#subtract function

diff2 = np.subtract(A,B)
print(diff2)
```

**multiplying a matrix by a scalar**

```python
x = 4
y = np.random.randint(89, size = (3,4))
print("Y=\n",y)
product = np.multiply(x,y)
print("4 * Y =  \n",product)
```

**multiplying two matrices**

```python
A = np.random.randint(8, size=(2,3))
print("A = \n", A)

B = np.random.randint(8, size = (3,4))
print("B =\n", B)

product = np.dot(A,B)
print("A * B = \n", product)
```
