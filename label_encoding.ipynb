#label encoding
Label Encoding is the process of converting categorical labels (text) into numeric values so that machine learning models can understand and process them.

For example:
 When to use Label Encoding
When the categorical feature is ordinal (has an order)

e.g., ['low', 'medium', 'high'] → [0, 1, 2]

When you want to quickly convert categories into numbers for models like Tree-based algorithms (e.g., Decision Trees, Random Forests)


Label Encoding can cause problems with non-ordinal categorical features.

Example:
['apple', 'banana', 'orange'] → [0, 1, 2]
A model might think "orange > banana > apple", which doesn't make sense.


```python
#importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder

```

##label encoding of breast cancer


```python
#loading the data from csv file to pandas dataframe
cancer_data =pd.read_csv('/breast_cancer_data_with_diagnosis.csv')
```

```python
#first five rows
cancer_data.head()
```

```python
#finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the lable encoder function to labeling these two stages of diagnosis
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 --> benign
1 --> malignant


```python
cancer_data['target'].value_counts()
```
