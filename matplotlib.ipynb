#matplotlib is used to plot data


```python
import matplotlib.pyplot as plt
```

```python
#as to plot data we need some data ,can be gather from numpy
import numpy as np
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

plotting the data

```python
#plotting sin wave
plt.plot(x,y)
plt.show
```

```python
#plotting cos wave
plt.plot(x,z)
plt.show()
```

```python
#adding title and labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sin wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'b+')
```

```python
plt.plot(x,y, 'g.')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x) ,'g*')
plt.plot(x,np.cos(x) , 'r^')

```

bar plot

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['eng', 'french' , 'latin', 'german']
people =[10, 45, 90, 44]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('peoples')
```

pie chart


```python
fig1= plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['eng', 'french' , 'latin', 'german']
people =[10, 45, 90, 44]
ax.pie(people, labels=languages,autopct ='%1.1f%%')
```

scatter plot


```python
x =np.linspace(0,10, 30)
y = np.sin(x)
z =np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
```

3d scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,c=z,cmap='Blues')
```
