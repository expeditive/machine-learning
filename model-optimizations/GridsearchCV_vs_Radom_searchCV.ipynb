# **Grid search CV**
Grid Search CV (Cross-Validation) is a powerful technique in machine learning used for hyperparameter tuning. It systematically tries every combination of parameters you specify and uses cross-validation to evaluate which combination performs best.

🔍 What Is Grid Search?
In most machine learning models, you have hyperparameters (like learning rate, number of trees, k in k-NN, etc.) that are not learned from the data but need to be set before training. Grid Search helps you automatically search through multiple combinations of these hyperparameters to find the best model.

✅ How Grid Search CV Works (Step-by-Step):
Define the model (e.g., SVM, RandomForest, etc.).

Create a parameter grid: A dictionary of hyperparameters and their possible values.

Set up cross-validation (like 5-fold, 10-fold).

For every combination of hyperparameters:

Split the dataset into training and validation sets multiple times (as per CV).

Train the model on the training part.

Evaluate the model on the validation part.

Record the average performance score.

Choose the best combination (with the highest average score).

```python
#loading the svc model
model = SVC()
```

```python
#hyperparameters

parameters = {
    'kernel':['linear','poly','rbf','sigmoid'] ,
    'C' : [1, 5, 10, 20]
}
```

```python
#grid  search
classifier = GridSearchCV(model , parameters, cv=5)
```

```python
#fitting the data to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
#best parameter

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#highest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
#loading the results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

Highest accuracy =  95.2%

best parameters = {'C': 10, 'kernel':'linear' }

# **Randomized search CV**

Sure! Let's dive into **Randomized Search CV** in detail — it's an efficient alternative to Grid Search for hyperparameter tuning.

---

## 🔍 What Is Randomized Search CV?

**Randomized Search CV** is a technique to find the best combination of hyperparameters for a machine learning model by **randomly sampling combinations from a defined range of values** instead of trying every single one (as in Grid Search).

It still uses **cross-validation (CV)** to evaluate the performance of each combination but is **faster and more efficient**, especially when you have a large number of hyperparameters or possible values.

---

## ⚙️ How It Works (Step-by-Step)

1. **Define the model** (e.g., SVM, RandomForest, etc.).
2. **Create a parameter distribution**: A dictionary of hyperparameters and a list or distribution of possible values.
3. Choose how many combinations (`n_iter`) to try.
4. For each randomly selected combination:

   * Use **cross-validation** to evaluate it.
   * Record the performance score.
5. **Return the best combination** based on the average CV score.

---

---

## 📈 Benefits of Randomized Search

* **Faster** than Grid Search when the number of hyperparameters or their values is large.
* **Can find a good solution quickly** with fewer evaluations.
* Allows using **distributions** (like normal, uniform, or integer ranges).

---

## ⚠️ Limitations

* May miss the best combination if it's not sampled.
* **Results vary** depending on the random seed and number of iterations.
* Not exhaustive like Grid Search.

---

## 🔁 When to Use What?

| Situation                        | Use                |
| -------------------------------- | ------------------ |
| Small search space               | GridSearchCV       |
| Large or continuous search space | RandomizedSearchCV |
| Fast results with tradeoff       | RandomizedSearchCV |
| High computational cost          | RandomizedSearchCV |
