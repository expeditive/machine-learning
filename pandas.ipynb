here we can not perfrom numeric operations like finding mean maximum minimum or so we have to convert the data into numeric to perform these operations

```python
#coverting categorical data to Numeric
boston_df["CHAS"] = pd.to_numeric(boston_df["CHAS"],errors = "coerce")
boston_df["RAD"] = pd.to_numeric(boston_df["RAD"],errors = "coerce")
```

```python
#calculating mean column wise
boston_df.mean()
```

```python
#calculating std deviation column wise
boston_df.std()
```

```python
#minimum value
boston_df.min()
```
