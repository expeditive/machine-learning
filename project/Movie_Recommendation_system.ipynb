##**difflib library**
 in Python is a standard library module used for comparing sequences, particularly strings. It helps you find the differences between two pieces of text or lists, generate human-readable comparisons, and even create tools like "diff" or "merge".

Key Features of difflib:
Compare Strings or Lists: It can compare two sequences (strings, lists, etc.) to find similarities and differences.

Show Differences: It can show changes in a format similar to Unix diff (like added/removed lines).

Find Close Matches: Useful in fuzzy string matching or spell-checking.

Get Matching Blocks: Finds subsequences that match between inputs.



## **TF-IDF Vectorizer**
 stands for Term Frequency‚ÄìInverse Document Frequency Vectorizer. It's a technique used in Natural Language Processing (NLP) to convert text data into numerical feature vectors, giving importance to unique and relevant words in a collection of documents.



# **Cosine Similarity**
is a metric used to measure how similar two vectors are, regardless of their magnitude. It‚Äôs widely used in text analysis, especially with TF-IDF vectors, to compare how similar two documents or sentences are.

üîç Intuition
Cosine similarity measures the cosine of the angle between two vectors in an n-dimensional space. A smaller angle (closer to 0¬∞) means higher similarity.



```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-processing

```python
#loading the data from  the csv file to pandas dataframe
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
#selecting the relevant features for recommenndation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with null string

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combining all the 5 selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

convering the text data to feature vectors

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

cosine similarity

```python
#getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting movie name from the user
movie_name = input('Enter Your Favourite Movie Name : ')
```

```python
#creating the list with all the movie name given in the dataset

list_of_all_titles = movies_data['title'].tolist()
```

```python
print(list_of_all_titles)
```

```python
#finding the closest match for the movie given by user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
#finding the index of the movie with title
index_of_the_movie = movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

similarity_score = list(enumerate(similarity[index_of_the_movie]))


‚úÖ What this line does:

üëâ This gives you a list of similarity scores between one movie (at index_of_the_movie) and all other movies.

üëâ This gives you pairs like (index, score), where:

score is how similar that movie is to the selected one.





```python
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on thier similarity score

sorted_similar_movie = sorted(similarity_score,key = lambda x:x[1], reverse=True)
print(sorted_similar_movie)
```

```python
#print the name of similar movies
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movie:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]

  if(i<30):
    print(i,'.', title_from_index)
    i+=1
```

# **Reccomendatation system**

```python
#getting movie name from the user
movie_name = input('Enter Your Favourite Movie Name : ')


#creating the list with all the movie name given in the dataset
list_of_all_titles = movies_data['title'].tolist()

#finding the closest match for the movie given by user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match=find_close_match[0]

#finding the index of the movie with title
index_of_the_movie = movies_data[movies_data.title==close_match]['index'].values[0]

# getting list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))

#sorting the movies based on thier similarity score
sorted_similar_movie = sorted(similarity_score,key = lambda x:x[1], reverse=True)

#print the name of similar movies
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movie:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]

  if(i<30):
    print(i,'.', title_from_index)
    i+=1
```
