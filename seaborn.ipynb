##seaborn
data visualisation library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

seaborn has some built in data set

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#visualize the data
sns.relplot(data=tips, x='total_bill',y='tip')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip', col='day',)
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',hue='smoker')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='smoker',hue='sex',)
```

```python
sns.relplot(data=tips,x='total_bill', y='tip',col='smoker',style='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',style='smoker',hue='sex',size='size')
```

```python
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='time',style='smoker',hue='sex',size='size')
```
